```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def comma_converter(x):
    if type(x) == str and ',' in x:
        a = x.split(',')
        return float(a[0] + '.' + a[1])
    else:
        return float(x)
    
#Converting the datatype 
df['distance'] = df['distance'].apply(comma_converter)
df['consume'] = df['consume'].apply(comma_converter)
df['temp_inside'] = df['temp_inside'].apply(comma_converter)
df['refill liters'] = df['refill liters'].apply(comma_converter)
```

```python
# adding a column which is the Liters consumed during the specified covered distance
df["consume/distance"] = (df["consume"]/100)*df["distance"]
```

```python
dist_avg_E10 = df[df['gas_type'] == 'E10']['distance'].mean()
dist_avg_SP98 = df[df['gas_type'] == 'SP98']['distance'].mean()

avg_consume_E10 = df[df['gas_type'] == 'E10']['consume/distance'].mean()
avg_consume_SP98 = df[df['gas_type'] == 'SP98']['consume/distance'].mean()

print("Average Distance Traveled with E10: ", round(dist_avg_E10,2),"km")
print("Average Distance Traveled with SP98: ", round(dist_avg_SP98,2),"km")

print("Average consumption per ride (E10): ", round(avg_consume_E10,2),"Liters")
print("Average consumption per ride (SP98): ", round(avg_consume_SP98,2),"Liters")
```

```python
dummies = pd.get_dummies(df['gas_type'])
df = pd.concat([df, dummies], axis=1)
```

```python
df["E10"] = df["E10"].replace({True:1,False:0})
```

```python
df["SP98"] = df["SP98"].replace({True:1,False:0})
```

```python
df = df.drop(columns=["specials","gas_type","refill liters","refill gas","consume"],axis=1)
```

```python
Y = df["consume/distance"]
df = df.drop(columns="consume/distance", axis=1)
```

```python
corr = df.corr()
matrix = np.triu(corr)
np.fill_diagonal(matrix, False)
sns.heatmap(corr, annot=True, fmt='.2f', mask=matrix)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df["temp_inside"] = df["temp_inside"].fillna(df["temp_inside"].mean())
```

```python
X_test, X_train, y_test, y_train = train_test_split(df,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
lm = LinearRegression()

model = lm.fit(X_train_scaled,y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
fig, ax = plt.subplots(1,3,figsize=(14,4))
ax[0].plot(y_pred, y_test, 'o')
ax[0].set_xlabel("y_test_transformed")
ax[0].set_ylabel("y_pred")
ax[0].set_title("Test Set -Predicted vs real")

# Get a histogram of the residuals ie: y - y_pred.  Homoscedasticity
# Does it resemble a normal distribution?
ax[1].hist(y_test - y_pred)
ax[1].set_xlabel("Test y-y_pred")
ax[1].set_title("Test Set Residual histogram")

ax[2].plot(y_pred,y_test- y_pred,"o")
ax[2].set_xlabel("predicted")
ax[2].set_ylabel("residuals")
ax[2].set_title("Residuals by Predicted")
ax[2].plot(y_pred,np.zeros(len(y_pred)),linestyle='dashed')
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse, r2_score as r2
mse = mse(y_test,y_pred)
print("MSE =",mse)
```

```python
mae = mae(y_test,y_pred)
print("MAE =",mae)
```

```python
import math
RMSE = math.sqrt(mse)
print("RMSE =",RMSE)
```

```python
r2 = r2(y_test,y_pred)
print("R2 =",r2)
```
